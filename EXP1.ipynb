```python
import numpy as np
data=[int(i) for i in input().split()]
Max=max(data)
lenth=len(data)
x=[]
freq=[]
for i in range(Max+1):
  c=0
  for j in range(lenth):
    if data[j]==i:
      c+=1
  x.append(i)
  freq.append(c)
sf=np.sum(freq)
p=[freq[i]/sf for i in range(Max+1)]
mean=np.inner(x,p)
ex2=np.inner(np.square(x),p)
var=ex2-mean**2
std=np.sqrt(var)
print(f"X\tP(x)")
for i in range(Max+1):
  if freq[i]>0:
    print(f'{x[i]}\t{p[i]:.3f} ')
print(f"mean {mean}")
print(f'var {var}')
print(f"std {std}")
```
